# Restaurants

```python
restaurants.shape
```

```python
restaurants.drop('address', axis=1, inplace=True)
```

```python
# restaurants name column - cleaning
restaurants = restaurants[restaurants['name'].str.split().str.len() <= 5]
restaurants = restaurants[~restaurants['name'].str.contains(r'\d', regex=True, na=False)]
restaurants['name'] = (restaurants['name'].str.strip().str.replace(r'[^a-zA-Z\s]', '', regex=True))
restaurants = restaurants[restaurants['name'].str.len() > 3]
```

```python
words = ['food', 'service', 'taste', 'ambience', 'experience', 'amazing', 'good', 'great', 'must', 'try', 'recommended', 'delicious',
'rating', 'rated', 'review', 'cake', 'Alley', 'Recommend', 'Goa 0 Km', 'Bohra', 'Foreign', 'Bati', 'Betroot', 'Boti Kabab', 'Butter naan'
'fish', 'Dal', 'fritters', 'balls', 'sausages', 'tomato', 'omllete', 'Cheese burst pizza', 'Dosas', 'soup', 'nice', 'fish', 'bread',
'I may not visit again', 'apple', 'karimeen fry', 'Jal jeerra', 'Kadi', 'Lachha Paratha', 'Fried Hanbanero Chicken burger', 'Meals',
'Luchi Cholar Day', 'Italian on Tuesdays', 'Lamb Racks', 'Minted Arancini', 'Mixed veg subzi', '√É', 'Chop', 'Fry', 'prepared', 'supreme',
'duplex', 'gobi', 'peri', 'stay', 'mushrooms', 'thai', 'tandoori', 'drinks', 'tikka', 'sweet', 'paneer', 'starters', 'fungi', 'go',
'say', 'and', 'awesome', 'style', 'potato', 'me', 'fun', 'but', 'cheap', 'chicken', 'colorful', 'crispy', 'quick', 'fries', 'thank',
'fresh', 'you', 'little', 'it', 'wheat', 'money', 'veggies', 'ice-cream', 'quantity', 'totally', 'though', 'tried', 'the', 'thanks',
'well', 'left', 'staffs', 'spinach', 'no', 'sabzi', 'salads', 'screening', 'sizzlers', 'roti', 'chilli', 'place', 'work', 'of', 'not',
'bowl', 'advice', 'mutton', 'music', 'mushroom', 'main', 'it', 'is', 'check', 'chic', 'cocktail', 'leaves', 'bar', 'Rubs', 'compact',
'fried', 'staff', 'busy', 'ghee', 'guava', 'chocolate', 'people', 'had', 'continental', 'how', 'however', 'lights', 'was', 'beef',
'tea', 'lemon', 'chips', 'much', 'were', 'olives', 'rice', 'cabbage', 'pastas']

restaurants = restaurants[~restaurants['name'].str.contains('|'.join(words), case=False, na=False)]
```

```python
# online_order column and book_table column
restaurants = restaurants[(restaurants['online_order'] == 'Yes') | (restaurants['online_order'] == 'No')]
restaurants = restaurants[(restaurants['book_table'] == 'Yes') | (restaurants['book_table'] == 'No')]
```

```python
# rate column
restaurants['rate'] = restaurants['rate'].str.split('/').str.get(0)
```

```python
restaurants['rate'] = restaurants['rate'].replace(['NEW', '-'], 0)
restaurants['rate'] = pd.to_numeric(restaurants['rate'].str.strip())
restaurants['rate'] = restaurants['rate'].fillna(0)
```

```python
# location column
restaurants = restaurants.dropna(subset=['location'])
```

```python
# approx_cost(for two people) column
restaurants['approx_cost(for two people)'] = restaurants['approx_cost(for two people)'].fillna(restaurants['approx_cost(for two people)'].mean())
```

```python
# rest_type column
restaurants['rest_type'] = restaurants['rest_type'].str.split(',').str[0].str.strip()
restaurants['rest_type'] = restaurants['rest_type'].fillna('Unknown')
```

```python
# listed_in(type) column
restaurants['listed_in(type)'] = restaurants['listed_in(type)'].fillna('Unknown')
```

```python
# dish_liked column
restaurants['dish_liked'] = restaurants['dish_liked'].str.split(',').str.get(0)
restaurants['dish_liked'] = restaurants['dish_liked'].fillna('Unknown')
```

```python
# cuisines
restaurants['cuisines'] = restaurants['cuisines'].str.split(',').str.get(0)
restaurants['cuisines'] = restaurants['cuisines'].fillna('Unknown')
```

```python
# listed_in(city)
restaurants['listed_in(city)'] = restaurants['listed_in(city)'].fillna(restaurants['location'])
```

```python
def clean_location(x):
    x = str(x)

    if "Whitefield" in x:
        return "Whitefield"

    elif "Koramangala" in x:
        return "Koramangala"

    else:
        return x
```

```python
restaurants['listed_in(city)'] = restaurants['listed_in(city)'].apply(clean_location)
```

```python
# votes
restaurants['votes'] = restaurants['votes'].astype('float64')
```

```python
restaurants.rename(columns={
    'RestaurantID': 'restaurant_id',
    'name': 'restaurant_name',
    'online_order': 'online_order',
    'book_table': 'table_booking',
    'rate': 'rating',
    'votes': 'total_votes',
    'location': 'location',
    'rest_type': 'restaurant_type',
    'dish_liked': 'dishes_liked',
    'cuisines': 'cuisines',
    'approx_cost(for two people)': 'cost_for_two',
    'listed_in(type)': 'service_type',
    'listed_in(city)': 'city_category'
}, inplace=True)
```

```python
restaurants.to_csv('restaurants.csv', index=False)
```
